```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# 1. Download the Dataset
url = 'https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data'
columns = ['SepalLength', 'SepalWidth', 'PetalLength', 'PetalWidth', 'Species']
df = pd.read_csv(url, header=None, names=columns)
```

```python
# Inspect the DataFrame structure
print("Dataset Head:")
print(df.head())
```

```python
# 2. Feature Analysis
# Get the number of features and their types
feature_types = df.dtypes
numeric_features = df.select_dtypes(include=['float64', 'int64']).columns
nominal_features = df.select_dtypes(include=['object']).columns

print(f"\nNumeric Features: {numeric_features}")
print(f"Nominal Features: {nominal_features}")
```

```python
# 3. Compute Summary Statistics
# Compute summary statistics for numeric features
summary_statistics = df[numeric_features].describe()
print("\nSummary Statistics:")
print(summary_statistics)
```

```python
# 4. Data Visualization: Histograms
# Create a histogram for each numeric feature
df[numeric_features].hist(bins=20, figsize=(12, 8), edgecolor='black')
```

```python
# Customize the histogram appearance
plt.suptitle('Histograms of Numeric Features', fontsize=16)
plt.tight_layout()
plt.show()
```

```python
# 5. Data Visualization: Boxplots
# Create a boxplot for each numeric feature
plt.figure(figsize=(12, 8))
sns.boxplot(data=df[numeric_features])
```

```python
# Add labels and title to improve clarity
plt.title('Boxplots of Numeric Features', fontsize=16)
plt.xlabel('Features', fontsize=12)
plt.ylabel('Values', fontsize=12)
```

```python
# Show the plot
plt.tight_layout()
plt.show()
```
